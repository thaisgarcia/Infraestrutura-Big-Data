```python
# 1. Definimos uma string multilinha com o conteúdo que queremos analisar
texto = """big data e o futuro da tecnologia
hadoop processa big data em clusters
spark e mais rapido que hadoop na memoria
big data com spark hadoop e python
machine learning usa big data para aprender"""

# 2. Criamos um arquivo chamado 'texto.txt' e escrevemos o conteúdo nele
with open("texto.txt", "w") as f:
    f.write(texto)

print("Arquivo 'texto.txt' criado com sucesso!")
```

```python
# 2. Definição do Script MapReduce (Extraindo a 3ª palavra)
codigo = r'''from mrjob.job import MRJob
import re

class WordCount(MRJob):
    def mapper(self, _, line):
        palavras = line.split()
        if len(palavras) >= 3:
            chave = re.sub(r'\W+', '', palavras[2].lower())
            if chave:
                yield chave, 1

    def reducer(self, palavra, contagens):
        yield palavra, sum(contagens)

if __name__ == "__main__":
    WordCount.run()
'''

with open("wordcount.py", "w") as f:
    f.write(codigo)
print("Script 'wordcount.py' configurado.")
```

```python
%%writefile loganalyzer.py
from mrjob.job import MRJob

class LogAnalyzer(MRJob):
    # O Mapper extrai a 3a palavra (indice 2) de cada linha
    def mapper(self, _, line):
        partes = line.split()
        if len(partes) >= 3:
            nivel = partes[2].upper()
            if nivel in ['ERROR', 'WARN', 'INFO']:
                yield nivel, 1

    # O Reducer soma as ocorrencias para cada nivel
    def reducer(self, nivel, contagens):
        yield nivel, sum(contagens)

if __name__ == '__main__':
    LogAnalyzer.run()
```

```python
logs_exemplo = """2023-10-21 10:00:01 INFO Sistema iniciado
2023-10-21 10:05:20 ERROR Falha na conexão com banco de dados
2023-10-21 10:10:00 INFO Usuário 'admin' logado
2023-10-21 10:15:45 WARN Uso de CPU acima de 80%
2023-10-21 10:20:10 ERROR Erro crítico no servidor de aplicação
2023-10-21 10:25:00 INFO Backup finalizado com sucesso
"""

with open("exemplo_logs.txt", "w") as f:
    f.write(logs_exemplo)

print("Arquivo 'exemplo_logs.txt' criado com sucesso!")
```
